```python
#Imported necessary packages and read data


import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
from sklearn.tree import DecisionTreeRegressor
from sklearn.tree import export_graphviz
from sklearn.ensemble import RandomForestRegressor

alcohol=pd.read_csv('alcohol_CDC.csv')
alcohol.head()
```

```python
#Descriptive analytics


alcohol.describe()
```

```python
#Code blocks for predictive analytics models; First block is for all states, subsequent code blocks for individual states; R^2 Values printed below


alcohol['STATE']=alcohol['STATE'].astype(str)
alcohol['GENDER']=alcohol['GENDER'].astype(str)
alcohol['ETHNICITY']=alcohol['ETHNICITY'].astype(str)
alcohol['GRADE']=alcohol['GRADE'].astype(str)
alcohol['STRAT TYPE']=alcohol['STRAT TYPE'].astype(str)
X,y=alcohol.iloc[:,:-1],alcohol['HIGH RISK PROPORTION']
X_dummies=pd.get_dummies(X)
X_train,X_test,y_train,y_test=train_test_split(X_dummies,y,random_state=0)
lin_alcohol=LinearRegression()
lin_alcohol.fit(X_train,y_train)
knn_alcohol=KNeighborsRegressor(n_neighbors=300)
knn_alcohol.fit(X_train,y_train)
lasso_alcohol=Lasso(alpha=0.01)
lasso_alcohol.fit(X_train,y_train)
ridge_alcohol=Ridge(alpha=0.01)
ridge_alcohol.fit(X_train,y_train)
dt_alcohol=DecisionTreeRegressor(random_state=0,max_leaf_nodes=10)
dt_alcohol.fit(X_train,y_train)
rf_alcohol=RandomForestRegressor(random_state=0,n_estimators=250)
rf_alcohol.fit(X_train,y_train)
export_graphviz(dt_alcohol,out_file='project_vis.dot',class_names=dt_alcohol.classes_,
               feature_names=X_dummies.columns,impurity=True,filled=False)
print('ALL STATES (21)')
print('Linear R^2 on Train Data:{:.3f}'.format(lin_alcohol.score(X_train,y_train)))
print('Linear R^2 on Test Data:{:.3f}'.format(lin_alcohol.score(X_test,y_test)))
print('KNN R^2 on Train Data:{:.3f}'.format(knn_alcohol.score(X_train,y_train)))
print('KNN R^2 on Test Data:{:.3f}'.format(knn_alcohol.score(X_test,y_test)))
print('Lasso R^2 on Train Data:{:.3f}'.format(lasso_alcohol.score(X_train,y_train)))
print('Lasso R^2 on Test Data:{:.3f}'.format(lasso_alcohol.score(X_test,y_test)))
print('Ridge R^2 on Train Data:{:.3f}'.format(ridge_alcohol.score(X_train,y_train)))
print('Ridge R^2 on Test Data:{:.3f}'.format(ridge_alcohol.score(X_test,y_test)))
print('Decision Tree R^2 on Train Data:{:.3f}'.format(dt_alcohol.score(X_train,y_train)))
print('Decision Tree R^2 on Test Data:{:.3f}'.format(dt_alcohol.score(X_test,y_test)))
print('Random Forest Regressor R^2 on Train Data:{:.3f}'.format(rf_alcohol.score(X_train,y_train)))
print('Random Forest Regressor R^2 on Test Data:{:.3f}'.format(rf_alcohol.score(X_test,y_test)))
print('')

AL=alcohol['STATE']=='AL'
Alabama=alcohol[AL]
X_AL,y_AL=Alabama.iloc[:,:-1],Alabama['HIGH RISK PROPORTION']
X_AL_dummies=pd.get_dummies(X_AL)
X_AL_train,X_AL_test,y_AL_train,y_AL_test=train_test_split(X_AL_dummies,y_AL,random_state=0)
lin_AL=LinearRegression()
lin_AL.fit(X_AL_train,y_AL_train)
knn_AL=KNeighborsRegressor(n_neighbors=3)
knn_AL.fit(X_AL_train,y_AL_train)
lasso_AL=Lasso(alpha=0.01)
lasso_AL.fit(X_AL_train,y_AL_train)
ridge_AL=Ridge(alpha=0.01)
ridge_AL.fit(X_AL_train,y_AL_train)
dt_AL=DecisionTreeRegressor(random_state=0,max_leaf_nodes=2)
dt_AL.fit(X_AL_train,y_AL_train)
rf_AL=RandomForestRegressor(random_state=0,n_estimators=100)
rf_AL.fit(X_AL_train,y_AL_train)
print('ALABAMA')
print('AL Linear R^2 on Train Data:{:.3f}'.format(lin_AL.score(X_AL_train,y_AL_train)))
print('AL Linear R^2 on Test Data:{:.3f}'.format(lin_AL.score(X_AL_test,y_AL_test)))
print('AL KNN R^2 on Train Data:{:.3f}'.format(knn_AL.score(X_AL_train,y_AL_train)))
print('AL KNN R^2 on Test Data:{:.3f}'.format(knn_AL.score(X_AL_test,y_AL_test)))
print('AL Lasso R^2 on Train Data:{:.3f}'.format(lasso_AL.score(X_AL_train,y_AL_train)))
print('AL Lasso R^2 on Test Data:{:.3f}'.format(lasso_AL.score(X_AL_test,y_AL_test)))
print('AL Ridge R^2 on Train Data:{:.3f}'.format(ridge_AL.score(X_AL_train,y_AL_train)))
print('AL Ridge R^2 on Test Data:{:.3f}'.format(ridge_AL.score(X_AL_test,y_AL_test)))
print('AL Decision Tree R^2 on Train Data:{:.3f}'.format(dt_AL.score(X_AL_train,y_AL_train)))
print('AL Decision Tree R^2 on Test Data:{:.3f}'.format(dt_AL.score(X_AL_test,y_AL_test)))
print('AL Random Forest Regressor R^2 on Train Data:{:.3f}'.format(rf_AL.score(X_AL_train,y_AL_train)))
print('AL Random Forest Regressor R^2 on Test Data:{:.3f}'.format(rf_AL.score(X_AL_test,y_AL_test)))
print('')

CO=alcohol['STATE']=='CO'
Colorado=alcohol[CO]
X_CO,y_CO=Colorado.iloc[:,:-1],Colorado['HIGH RISK PROPORTION']
X_CO_dummies=pd.get_dummies(X_CO)
X_CO_train,X_CO_test,y_CO_train,y_CO_test=train_test_split(X_CO_dummies,y_CO,random_state=0)
lin_CO=LinearRegression()
lin_CO.fit(X_CO_train,y_CO_train)
knn_CO=KNeighborsRegressor(n_neighbors=3)
knn_CO.fit(X_CO_train,y_CO_train)
lasso_CO=Lasso(alpha=0.01)
lasso_CO.fit(X_CO_train,y_CO_train)
ridge_CO=Ridge(alpha=0.01)
ridge_CO.fit(X_CO_train,y_CO_train)
dt_CO=DecisionTreeRegressor(random_state=0,max_leaf_nodes=2)
dt_CO.fit(X_CO_train,y_CO_train)
rf_CO=RandomForestRegressor(random_state=0,n_estimators=100)
rf_CO.fit(X_CO_train,y_CO_train)
print('COLORADO')
print('CO Linear R^2 on Train Data:{:.3f}'.format(lin_CO.score(X_CO_train,y_CO_train)))
print('CO Linear R^2 on Test Data:{:.3f}'.format(lin_CO.score(X_CO_test,y_CO_test)))
print('CO KNN R^2 on Train Data:{:.3f}'.format(knn_CO.score(X_CO_train,y_CO_train)))
print('CO KNN R^2 on Test Data:{:.3f}'.format(knn_CO.score(X_CO_test,y_CO_test)))
print('CO Lasso R^2 on Train Data:{:.3f}'.format(lasso_CO.score(X_CO_train,y_CO_train)))
print('CO Lasso R^2 on Test Data:{:.3f}'.format(lasso_CO.score(X_CO_test,y_CO_test)))
print('CO Ridge R^2 on Train Data:{:.3f}'.format(ridge_CO.score(X_CO_train,y_CO_train)))
print('CO Ridge R^2 on Test Data:{:.3f}'.format(ridge_CO.score(X_CO_test,y_CO_test)))
print('CO Decision Tree R^2 on Train Data:{:.3f}'.format(dt_CO.score(X_CO_train,y_CO_train)))
print('CO Decision Tree R^2 on Test Data:{:.3f}'.format(dt_CO.score(X_CO_test,y_CO_test)))
print('CO Random Forest Regressor R^2 on Train Data:{:.3f}'.format(rf_CO.score(X_CO_train,y_CO_train)))
print('CO Random Forest Regressor R^2 on Test Data:{:.3f}'.format(rf_CO.score(X_CO_test,y_CO_test)))
print('')

DE=alcohol['STATE']=='DE'
Delaware=alcohol[DE]
X_DE,y_DE=Delaware.iloc[:,:-1],Delaware['HIGH RISK PROPORTION']
X_DE_dummies=pd.get_dummies(X_DE)
X_DE_train,X_DE_test,y_DE_train,y_DE_test=train_test_split(X_DE_dummies,y_DE,random_state=0)
lin_DE=LinearRegression()
lin_DE.fit(X_DE_train,y_DE_train)
knn_DE=KNeighborsRegressor(n_neighbors=3)
knn_DE.fit(X_DE_train,y_DE_train)
lasso_DE=Lasso(alpha=0.01)
lasso_DE.fit(X_DE_train,y_DE_train)
ridge_DE=Ridge(alpha=0.01)
ridge_DE.fit(X_DE_train,y_DE_train)
dt_DE=DecisionTreeRegressor(random_state=0,max_leaf_nodes=2)
dt_DE.fit(X_DE_train,y_DE_train)
rf_DE=RandomForestRegressor(random_state=0,n_estimators=100)
rf_DE.fit(X_DE_train,y_DE_train)
print('DELAWARE')
print('DE Linear R^2 on Train Data:{:.3f}'.format(lin_DE.score(X_DE_train,y_DE_train)))
print('DE Linear R^2 on Test Data:{:.3f}'.format(lin_DE.score(X_DE_test,y_DE_test)))
print('DE KNN R^2 on Train Data:{:.3f}'.format(knn_DE.score(X_DE_train,y_DE_train)))
print('DE KNN R^2 on Test Data:{:.3f}'.format(knn_DE.score(X_DE_test,y_DE_test)))
print('DE Lasso R^2 on Train Data:{:.3f}'.format(lasso_DE.score(X_DE_train,y_DE_train)))
print('DE Lasso R^2 on Test Data:{:.3f}'.format(lasso_DE.score(X_DE_test,y_DE_test)))
print('DE Ridge R^2 on Train Data:{:.3f}'.format(ridge_DE.score(X_DE_train,y_DE_train)))
print('DE Ridge R^2 on Test Data:{:.3f}'.format(ridge_DE.score(X_DE_test,y_DE_test)))
print('DE Decision Tree R^2 on Train Data:{:.3f}'.format(dt_DE.score(X_DE_train,y_DE_train)))
print('DE Decision Tree R^2 on Test Data:{:.3f}'.format(dt_DE.score(X_DE_test,y_DE_test)))
print('DE Random Forest Regressor R^2 on Train Data:{:.3f}'.format(rf_DE.score(X_DE_train,y_DE_train)))
print('DE Random Forest Regressor R^2 on Test Data:{:.3f}'.format(rf_DE.score(X_DE_test,y_DE_test)))
print('')

FL=alcohol['STATE']=='FL'
Florida=alcohol[FL]
X_FL,y_FL=Florida.iloc[:,:-1],Florida['HIGH RISK PROPORTION']
X_FL_dummies=pd.get_dummies(X_FL)
X_FL_train,X_FL_test,y_FL_train,y_FL_test=train_test_split(X_FL_dummies,y_FL,random_state=0)
lin_FL=LinearRegression()
lin_FL.fit(X_FL_train,y_FL_train)
knn_FL=KNeighborsRegressor(n_neighbors=3)
knn_FL.fit(X_FL_train,y_FL_train)
lasso_FL=Lasso(alpha=0.01)
lasso_FL.fit(X_FL_train,y_FL_train)
ridge_FL=Ridge(alpha=0.01)
ridge_FL.fit(X_FL_train,y_FL_train)
dt_FL=DecisionTreeRegressor(random_state=0,max_leaf_nodes=2)
dt_FL.fit(X_FL_train,y_FL_train)
rf_FL=RandomForestRegressor(random_state=0,n_estimators=100)
rf_FL.fit(X_FL_train,y_FL_train)
print('FLORIDA')
print('FL Linear R^2 on Train Data:{:.3f}'.format(lin_FL.score(X_FL_train,y_FL_train)))
print('FL Linear R^2 on Test Data:{:.3f}'.format(lin_FL.score(X_FL_test,y_FL_test)))
print('FL KNN R^2 on Train Data:{:.3f}'.format(knn_FL.score(X_FL_train,y_FL_train)))
print('FL KNN R^2 on Test Data:{:.3f}'.format(knn_FL.score(X_FL_test,y_FL_test)))
print('FL Lasso R^2 on Train Data:{:.3f}'.format(lasso_FL.score(X_FL_train,y_FL_train)))
print('FL Lasso R^2 on Test Data:{:.3f}'.format(lasso_FL.score(X_FL_test,y_FL_test)))
print('FL Ridge R^2 on Train Data:{:.3f}'.format(ridge_FL.score(X_FL_train,y_FL_train)))
print('FL Ridge R^2 on Test Data:{:.3f}'.format(ridge_FL.score(X_FL_test,y_FL_test)))
print('FL Decision Tree R^2 on Train Data:{:.3f}'.format(dt_FL.score(X_FL_train,y_FL_train)))
print('FL Decision Tree R^2 on Test Data:{:.3f}'.format(dt_FL.score(X_FL_test,y_FL_test)))
print('FL Random Forest Regressor R^2 on Train Data:{:.3f}'.format(rf_FL.score(X_FL_train,y_FL_train)))
print('FL Random Forest Regressor R^2 on Test Data:{:.3f}'.format(rf_FL.score(X_FL_test,y_FL_test)))
print('')

GA=alcohol['STATE']=='GA'
Georgia=alcohol[GA]
X_GA,y_GA=Georgia.iloc[:,:-1],Georgia['HIGH RISK PROPORTION']
X_GA_dummies=pd.get_dummies(X_GA)
X_GA_train,X_GA_test,y_GA_train,y_GA_test=train_test_split(X_GA_dummies,y_GA,random_state=0)
lin_GA=LinearRegression()
lin_GA.fit(X_GA_train,y_GA_train)
knn_GA=KNeighborsRegressor(n_neighbors=3)
knn_GA.fit(X_GA_train,y_GA_train)
lasso_GA=Lasso(alpha=0.01)
lasso_GA.fit(X_GA_train,y_GA_train)
ridge_GA=Ridge(alpha=0.01)
ridge_GA.fit(X_GA_train,y_GA_train)
dt_GA=DecisionTreeRegressor(random_state=0,max_leaf_nodes=2)
dt_GA.fit(X_GA_train,y_GA_train)
rf_GA=RandomForestRegressor(random_state=0,n_estimators=100)
rf_GA.fit(X_GA_train,y_GA_train)
print('GEORGIA')
print('GA Linear R^2 on Train Data:{:.3f}'.format(lin_GA.score(X_GA_train,y_GA_train)))
print('GA Linear R^2 on Test Data:{:.3f}'.format(lin_GA.score(X_GA_test,y_GA_test)))
print('GA KNN R^2 on Train Data:{:.3f}'.format(knn_GA.score(X_GA_train,y_GA_train)))
print('GA KNN R^2 on Test Data:{:.3f}'.format(knn_GA.score(X_GA_test,y_GA_test)))
print('GA Lasso R^2 on Train Data:{:.3f}'.format(lasso_GA.score(X_GA_train,y_GA_train)))
print('GA Lasso R^2 on Test Data:{:.3f}'.format(lasso_GA.score(X_GA_test,y_GA_test)))
print('GA Ridge R^2 on Train Data:{:.3f}'.format(ridge_GA.score(X_GA_train,y_GA_train)))
print('GA Ridge R^2 on Test Data:{:.3f}'.format(ridge_GA.score(X_GA_test,y_GA_test)))
print('GA Decision Tree R^2 on Train Data:{:.3f}'.format(dt_GA.score(X_GA_train,y_GA_train)))
print('GA Decision Tree R^2 on Test Data:{:.3f}'.format(dt_GA.score(X_GA_test,y_GA_test)))
print('GA Random Forest Regressor R^2 on Train Data:{:.3f}'.format(rf_GA.score(X_GA_train,y_GA_train)))
print('GA Random Forest Regressor R^2 on Test Data:{:.3f}'.format(rf_GA.score(X_GA_test,y_GA_test)))
print('')

HI=alcohol['STATE']=='HI'
Hawaii=alcohol[HI]
X_HI,y_HI=Hawaii.iloc[:,:-1],Hawaii['HIGH RISK PROPORTION']
X_HI_dummies=pd.get_dummies(X_HI)
X_HI_train,X_HI_test,y_HI_train,y_HI_test=train_test_split(X_HI_dummies,y_HI,random_state=0)
lin_HI=LinearRegression()
lin_HI.fit(X_HI_train,y_HI_train)
knn_HI=KNeighborsRegressor(n_neighbors=3)
knn_HI.fit(X_HI_train,y_HI_train)
lasso_HI=Lasso(alpha=0.01)
lasso_HI.fit(X_HI_train,y_HI_train)
ridge_HI=Ridge(alpha=0.01)
ridge_HI.fit(X_HI_train,y_HI_train)
dt_HI=DecisionTreeRegressor(random_state=0,max_leaf_nodes=2)
dt_HI.fit(X_HI_train,y_HI_train)
rf_HI=RandomForestRegressor(random_state=0,n_estimators=100)
rf_HI.fit(X_HI_train,y_HI_train)
print('HAWAII')
print('HI Linear R^2 on Train Data:{:.3f}'.format(lin_HI.score(X_HI_train,y_HI_train)))
print('HI Linear R^2 on Test Data:{:.3f}'.format(lin_HI.score(X_HI_test,y_HI_test)))
print('HI KNN R^2 on Train Data:{:.3f}'.format(knn_HI.score(X_HI_train,y_HI_train)))
print('HI KNN R^2 on Test Data:{:.3f}'.format(knn_HI.score(X_HI_test,y_HI_test)))
print('HI Lasso R^2 on Train Data:{:.3f}'.format(lasso_HI.score(X_HI_train,y_HI_train)))
print('HI Lasso R^2 on Test Data:{:.3f}'.format(lasso_HI.score(X_HI_test,y_HI_test)))
print('HI Ridge R^2 on Train Data:{:.3f}'.format(ridge_HI.score(X_HI_train,y_HI_train)))
print('HI Ridge R^2 on Test Data:{:.3f}'.format(ridge_HI.score(X_HI_test,y_HI_test)))
print('HI Decision Tree R^2 on Train Data:{:.3f}'.format(dt_HI.score(X_HI_train,y_HI_train)))
print('HI Decision Tree R^2 on Test Data:{:.3f}'.format(dt_HI.score(X_HI_test,y_HI_test)))
print('HI Random Forest Regressor R^2 on Train Data:{:.3f}'.format(rf_HI.score(X_HI_train,y_HI_train)))
print('HI Random Forest Regressor R^2 on Test Data:{:.3f}'.format(rf_HI.score(X_HI_test,y_HI_test)))
print('')

KY=alcohol['STATE']=='KY'
Kentucky=alcohol[KY]
X_KY,y_KY=Kentucky.iloc[:,:-1],Kentucky['HIGH RISK PROPORTION']
X_KY_dummies=pd.get_dummies(X_KY)
X_KY_train,X_KY_test,y_KY_train,y_KY_test=train_test_split(X_KY_dummies,y_KY,random_state=0)
lin_KY=LinearRegression()
lin_KY.fit(X_KY_train,y_KY_train)
knn_KY=KNeighborsRegressor(n_neighbors=3)
knn_KY.fit(X_KY_train,y_KY_train)
lasso_KY=Lasso(alpha=0.01)
lasso_KY.fit(X_KY_train,y_KY_train)
ridge_KY=Ridge(alpha=0.01)
ridge_KY.fit(X_KY_train,y_KY_train)
dt_KY=DecisionTreeRegressor(random_state=0,max_leaf_nodes=2)
dt_KY.fit(X_KY_train,y_KY_train)
rf_KY=RandomForestRegressor(random_state=0,n_estimators=100)
rf_KY.fit(X_KY_train,y_KY_train)
print('KENTUCKY')
print('KY Linear R^2 on Train Data:{:.3f}'.format(lin_KY.score(X_KY_train,y_KY_train)))
print('KY Linear R^2 on Test Data:{:.3f}'.format(lin_KY.score(X_KY_test,y_KY_test)))
print('KY KNN R^2 on Train Data:{:.3f}'.format(knn_KY.score(X_KY_train,y_KY_train)))
print('KY KNN R^2 on Test Data:{:.3f}'.format(knn_KY.score(X_KY_test,y_KY_test)))
print('KY Lasso R^2 on Train Data:{:.3f}'.format(lasso_KY.score(X_KY_train,y_KY_train)))
print('KY Lasso R^2 on Test Data:{:.3f}'.format(lasso_KY.score(X_KY_test,y_KY_test)))
print('KY Ridge R^2 on Train Data:{:.3f}'.format(ridge_KY.score(X_KY_train,y_KY_train)))
print('KY Ridge R^2 on Test Data:{:.3f}'.format(ridge_KY.score(X_KY_test,y_KY_test)))
print('KY Decision Tree R^2 on Train Data:{:.3f}'.format(dt_KY.score(X_KY_train,y_KY_train)))
print('KY Decision Tree R^2 on Test Data:{:.3f}'.format(dt_KY.score(X_KY_test,y_KY_test)))
print('KY Random Forest Regressor R^2 on Train Data:{:.3f}'.format(rf_KY.score(X_KY_train,y_KY_train)))
print('KY Random Forest Regressor R^2 on Test Data:{:.3f}'.format(rf_KY.score(X_KY_test,y_KY_test)))
print('')

ME=alcohol['STATE']=='ME'
Maine=alcohol[ME]
X_ME,y_ME=Maine.iloc[:,:-1],Maine['HIGH RISK PROPORTION']
X_ME_dummies=pd.get_dummies(X_ME)
X_ME_train,X_ME_test,y_ME_train,y_ME_test=train_test_split(X_ME_dummies,y_ME,random_state=0)
lin_ME=LinearRegression()
lin_ME.fit(X_ME_train,y_ME_train)
knn_ME=KNeighborsRegressor(n_neighbors=3)
knn_ME.fit(X_ME_train,y_ME_train)
lasso_ME=Lasso(alpha=0.01)
lasso_ME.fit(X_ME_train,y_ME_train)
ridge_ME=Ridge(alpha=0.01)
ridge_ME.fit(X_ME_train,y_ME_train)
dt_ME=DecisionTreeRegressor(random_state=0,max_leaf_nodes=2)
dt_ME.fit(X_ME_train,y_ME_train)
rf_ME=RandomForestRegressor(random_state=0,n_estimators=100)
rf_ME.fit(X_ME_train,y_ME_train)
export_graphviz(dt_ME,out_file='project_ME_vis.dot',class_names=dt_ME.classes_,
               feature_names=X_ME_dummies.columns,impurity=True,filled=False)
print('MAINE')
print('ME Linear R^2 on Train Data:{:.3f}'.format(lin_ME.score(X_ME_train,y_ME_train)))
print('ME Linear R^2 on Test Data:{:.3f}'.format(lin_ME.score(X_ME_test,y_ME_test)))
print('ME KNN R^2 on Train Data:{:.3f}'.format(knn_ME.score(X_ME_train,y_ME_train)))
print('ME KNN R^2 on Test Data:{:.3f}'.format(knn_ME.score(X_ME_test,y_ME_test)))
print('ME Lasso R^2 on Train Data:{:.3f}'.format(lasso_ME.score(X_ME_train,y_ME_train)))
print('ME Lasso R^2 on Test Data:{:.3f}'.format(lasso_ME.score(X_ME_test,y_ME_test)))
print('ME Ridge R^2 on Train Data:{:.3f}'.format(ridge_ME.score(X_ME_train,y_ME_train)))
print('ME Ridge R^2 on Test Data:{:.3f}'.format(ridge_ME.score(X_ME_test,y_ME_test)))
print('ME Decision Tree R^2 on Train Data:{:.3f}'.format(dt_ME.score(X_ME_train,y_ME_train)))
print('ME Decision Tree R^2 on Test Data:{:.3f}'.format(dt_ME.score(X_ME_test,y_ME_test)))
print('ME Random Forest Regressor R^2 on Train Data:{:.3f}'.format(rf_ME.score(X_ME_train,y_ME_train)))
print('ME Random Forest Regressor R^2 on Test Data:{:.3f}'.format(rf_ME.score(X_ME_test,y_ME_test)))
print('')

MD=alcohol['STATE']=='MD'
Maryland=alcohol[MD]
X_MD,y_MD=Maryland.iloc[:,:-1],Maryland['HIGH RISK PROPORTION']
X_MD_dummies=pd.get_dummies(X_MD)
X_MD_train,X_MD_test,y_MD_train,y_MD_test=train_test_split(X_MD_dummies,y_MD,random_state=0)
lin_MD=LinearRegression()
lin_MD.fit(X_MD_train,y_MD_train)
knn_MD=KNeighborsRegressor(n_neighbors=3)
knn_MD.fit(X_MD_train,y_MD_train)
lasso_MD=Lasso(alpha=0.01)
lasso_MD.fit(X_MD_train,y_MD_train)
ridge_MD=Ridge(alpha=0.01)
ridge_MD.fit(X_MD_train,y_MD_train)
dt_MD=DecisionTreeRegressor(random_state=0,max_leaf_nodes=2)
dt_MD.fit(X_MD_train,y_MD_train)
rf_MD=RandomForestRegressor(random_state=0,n_estimators=100)
rf_MD.fit(X_MD_train,y_MD_train)
print('MARYLAND')
print('MD Linear R^2 on Train Data:{:.3f}'.format(lin_MD.score(X_MD_train,y_MD_train)))
print('MD Linear R^2 on Test Data:{:.3f}'.format(lin_MD.score(X_MD_test,y_MD_test)))
print('MD KNN R^2 on Train Data:{:.3f}'.format(knn_MD.score(X_MD_train,y_MD_train)))
print('MD KNN R^2 on Test Data:{:.3f}'.format(knn_MD.score(X_MD_test,y_MD_test)))
print('MD Lasso R^2 on Train Data:{:.3f}'.format(lasso_MD.score(X_MD_train,y_MD_train)))
print('MD Lasso R^2 on Test Data:{:.3f}'.format(lasso_MD.score(X_MD_test,y_MD_test)))
print('MD Ridge R^2 on Train Data:{:.3f}'.format(ridge_MD.score(X_MD_train,y_MD_train)))
print('MD Ridge R^2 on Test Data:{:.3f}'.format(ridge_MD.score(X_MD_test,y_MD_test)))
print('MD Decision Tree R^2 on Train Data:{:.3f}'.format(dt_MD.score(X_MD_train,y_MD_train)))
print('MD Decision Tree R^2 on Test Data:{:.3f}'.format(dt_MD.score(X_MD_test,y_MD_test)))
print('MD Random Forest Regressor R^2 on Train Data:{:.3f}'.format(rf_MD.score(X_MD_train,y_MD_train)))
print('MD Random Forest Regressor R^2 on Test Data:{:.3f}'.format(rf_MD.score(X_MD_test,y_MD_test)))
print('')

MS=alcohol['STATE']=='MS'
Mississippi=alcohol[MS]
X_MS,y_MS=Mississippi.iloc[:,:-1],Mississippi['HIGH RISK PROPORTION']
X_MS_dummies=pd.get_dummies(X_MS)
X_MS_train,X_MS_test,y_MS_train,y_MS_test=train_test_split(X_MS_dummies,y_MS,random_state=0)
lin_MS=LinearRegression()
lin_MS.fit(X_MS_train,y_MS_train)
knn_MS=KNeighborsRegressor(n_neighbors=3)
knn_MS.fit(X_MS_train,y_MS_train)
lasso_MS=Lasso(alpha=0.01)
lasso_MS.fit(X_MS_train,y_MS_train)
ridge_MS=Ridge(alpha=0.01)
ridge_MS.fit(X_MS_train,y_MS_train)
dt_MS=DecisionTreeRegressor(random_state=0,max_leaf_nodes=2)
dt_MS.fit(X_MS_train,y_MS_train)
rf_MS=RandomForestRegressor(random_state=0,n_estimators=100)
rf_MS.fit(X_MS_train,y_MS_train)
print('MISSISSIPPI')
print('MS Linear R^2 on Train Data:{:.3f}'.format(lin_MS.score(X_MS_train,y_MS_train)))
print('MS Linear R^2 on Test Data:{:.3f}'.format(lin_MS.score(X_MS_test,y_MS_test)))
print('MS KNN R^2 on Train Data:{:.3f}'.format(knn_MS.score(X_MS_train,y_MS_train)))
print('MS KNN R^2 on Test Data:{:.3f}'.format(knn_MS.score(X_MS_test,y_MS_test)))
print('MS Lasso R^2 on Train Data:{:.3f}'.format(lasso_MS.score(X_MS_train,y_MS_train)))
print('MS Lasso R^2 on Test Data:{:.3f}'.format(lasso_MS.score(X_MS_test,y_MS_test)))
print('MS Ridge R^2 on Train Data:{:.3f}'.format(ridge_MS.score(X_MS_train,y_MS_train)))
print('MS Ridge R^2 on Test Data:{:.3f}'.format(ridge_MS.score(X_MS_test,y_MS_test)))
print('MS Decision Tree R^2 on Train Data:{:.3f}'.format(dt_MS.score(X_MS_train,y_MS_train)))
print('MS Decision Tree R^2 on Test Data:{:.3f}'.format(dt_MS.score(X_MS_test,y_MS_test)))
print('MS Random Forest Regressor R^2 on Train Data:{:.3f}'.format(rf_MS.score(X_MS_train,y_MS_train)))
print('MS Random Forest Regressor R^2 on Test Data:{:.3f}'.format(rf_MS.score(X_MS_test,y_MS_test)))
print('')

NV=alcohol['STATE']=='NV'
Nevada=alcohol[NV]
X_NV,y_NV=Nevada.iloc[:,:-1],Nevada['HIGH RISK PROPORTION']
X_NV_dummies=pd.get_dummies(X_NV)
X_NV_train,X_NV_test,y_NV_train,y_NV_test=train_test_split(X_NV_dummies,y_NV,random_state=0)
lin_NV=LinearRegression()
lin_NV.fit(X_NV_train,y_NV_train)
knn_NV=KNeighborsRegressor(n_neighbors=3)
knn_NV.fit(X_NV_train,y_NV_train)
lasso_NV=Lasso(alpha=0.01)
lasso_NV.fit(X_NV_train,y_NV_train)
ridge_NV=Ridge(alpha=0.01)
ridge_NV.fit(X_NV_train,y_NV_train)
dt_NV=DecisionTreeRegressor(random_state=0,max_leaf_nodes=2)
dt_NV.fit(X_NV_train,y_NV_train)
rf_NV=RandomForestRegressor(random_state=0,n_estimators=100)
rf_NV.fit(X_NV_train,y_NV_train)
print('NEVADA')
print('NV Linear R^2 on Train Data:{:.3f}'.format(lin_NV.score(X_NV_train,y_NV_train)))
print('NV Linear R^2 on Test Data:{:.3f}'.format(lin_NV.score(X_NV_test,y_NV_test)))
print('NV KNN R^2 on Train Data:{:.3f}'.format(knn_NV.score(X_NV_train,y_NV_train)))
print('NV KNN R^2 on Test Data:{:.3f}'.format(knn_NV.score(X_NV_test,y_NV_test)))
print('NV Lasso R^2 on Train Data:{:.3f}'.format(lasso_NV.score(X_NV_train,y_NV_train)))
print('NV Lasso R^2 on Test Data:{:.3f}'.format(lasso_NV.score(X_NV_test,y_NV_test)))
print('NV Ridge R^2 on Train Data:{:.3f}'.format(ridge_NV.score(X_NV_train,y_NV_train)))
print('NV Ridge R^2 on Test Data:{:.3f}'.format(ridge_NV.score(X_NV_test,y_NV_test)))
print('NV Decision Tree R^2 on Train Data:{:.3f}'.format(dt_NV.score(X_NV_train,y_NV_train)))
print('NV Decision Tree R^2 on Test Data:{:.3f}'.format(dt_NV.score(X_NV_test,y_NV_test)))
print('NV Random Forest Regressor R^2 on Train Data:{:.3f}'.format(rf_NV.score(X_NV_train,y_NV_train)))
print('NV Random Forest Regressor R^2 on Test Data:{:.3f}'.format(rf_NV.score(X_NV_test,y_NV_test)))
print('')

NM=alcohol['STATE']=='NM'
New_Mexico=alcohol[NM]
X_NM,y_NM=New_Mexico.iloc[:,:-1],New_Mexico['HIGH RISK PROPORTION']
X_NM_dummies=pd.get_dummies(X_NM)
X_NM_train,X_NM_test,y_NM_train,y_NM_test=train_test_split(X_NM_dummies,y_NM,random_state=0)
lin_NM=LinearRegression()
lin_NM.fit(X_NM_train,y_NM_train)
knn_NM=KNeighborsRegressor(n_neighbors=3)
knn_NM.fit(X_NM_train,y_NM_train)
lasso_NM=Lasso(alpha=0.01)
lasso_NM.fit(X_NM_train,y_NM_train)
ridge_NM=Ridge(alpha=0.01)
ridge_NM.fit(X_NM_train,y_NM_train)
dt_NM=DecisionTreeRegressor(random_state=0,max_leaf_nodes=2)
dt_NM.fit(X_NM_train,y_NM_train)
rf_NM=RandomForestRegressor(random_state=0,n_estimators=100)
rf_NM.fit(X_NM_train,y_NM_train)
print('NEW MEXICO')
print('NM Linear R^2 on Train Data:{:.3f}'.format(lin_NM.score(X_NM_train,y_NM_train)))
print('NM Linear R^2 on Test Data:{:.3f}'.format(lin_NM.score(X_NM_test,y_NM_test)))
print('NM KNN R^2 on Train Data:{:.3f}'.format(knn_NM.score(X_NM_train,y_NM_train)))
print('NM KNN R^2 on Test Data:{:.3f}'.format(knn_NM.score(X_NM_test,y_NM_test)))
print('NM Lasso R^2 on Train Data:{:.3f}'.format(lasso_NM.score(X_NM_train,y_NM_train)))
print('NM Lasso R^2 on Test Data:{:.3f}'.format(lasso_NM.score(X_NM_test,y_NM_test)))
print('NM Ridge R^2 on Train Data:{:.3f}'.format(ridge_NM.score(X_NM_train,y_NM_train)))
print('NM Ridge R^2 on Test Data:{:.3f}'.format(ridge_NM.score(X_NM_test,y_NM_test)))
print('NM Decision Tree R^2 on Train Data:{:.3f}'.format(dt_NM.score(X_NM_train,y_NM_train)))
print('NM Decision Tree R^2 on Test Data:{:.3f}'.format(dt_NM.score(X_NM_test,y_NM_test)))
print('NM Random Forest Regressor R^2 on Train Data:{:.3f}'.format(rf_NM.score(X_NM_train,y_NM_train)))
print('NM Random Forest Regressor R^2 on Test Data:{:.3f}'.format(rf_NM.score(X_NM_test,y_NM_test)))
print('')

NC=alcohol['STATE']=='NC'
North_Carolina=alcohol[NC]
X_NC,y_NC=North_Carolina.iloc[:,:-1],North_Carolina['HIGH RISK PROPORTION']
X_NC_dummies=pd.get_dummies(X_NC)
X_NC_train,X_NC_test,y_NC_train,y_NC_test=train_test_split(X_NC_dummies,y_NC,random_state=0)
lin_NC=LinearRegression()
lin_NC.fit(X_NC_train,y_NC_train)
knn_NC=KNeighborsRegressor(n_neighbors=3)
knn_NC.fit(X_NC_train,y_NC_train)
lasso_NC=Lasso(alpha=0.01)
lasso_NC.fit(X_NC_train,y_NC_train)
ridge_NC=Ridge(alpha=0.01)
ridge_NC.fit(X_NC_train,y_NC_train)
dt_NC=DecisionTreeRegressor(random_state=0,max_leaf_nodes=2)
dt_NC.fit(X_NC_train,y_NC_train)
rf_NC=RandomForestRegressor(random_state=0,n_estimators=100)
rf_NC.fit(X_NC_train,y_NC_train)
print('NORTH CAROLINA')
print('NC Linear R^2 on Train Data:{:.3f}'.format(lin_NC.score(X_NC_train,y_NC_train)))
print('NC Linear R^2 on Test Data:{:.3f}'.format(lin_NC.score(X_NC_test,y_NC_test)))
print('NC KNN R^2 on Train Data:{:.3f}'.format(knn_NC.score(X_NC_train,y_NC_train)))
print('NC KNN R^2 on Test Data:{:.3f}'.format(knn_NC.score(X_NC_test,y_NC_test)))
print('NC Lasso R^2 on Train Data:{:.3f}'.format(lasso_NC.score(X_NC_train,y_NC_train)))
print('NC Lasso R^2 on Test Data:{:.3f}'.format(lasso_NC.score(X_NC_test,y_NC_test)))
print('NC Ridge R^2 on Train Data:{:.3f}'.format(ridge_NC.score(X_NC_train,y_NC_train)))
print('NC Ridge R^2 on Test Data:{:.3f}'.format(ridge_NC.score(X_NC_test,y_NC_test)))
print('NC Decision Tree R^2 on Train Data:{:.3f}'.format(dt_NC.score(X_NC_train,y_NC_train)))
print('NC Decision Tree R^2 on Test Data:{:.3f}'.format(dt_NC.score(X_NC_test,y_NC_test)))
print('NC Random Forest Regressor R^2 on Train Data:{:.3f}'.format(rf_NC.score(X_NC_train,y_NC_train)))
print('NC Random Forest Regressor R^2 on Test Data:{:.3f}'.format(rf_NC.score(X_NC_test,y_NC_test)))
print('')

ND=alcohol['STATE']=='ND'
North_Dakota=alcohol[ND]
X_ND,y_ND=North_Dakota.iloc[:,:-1],North_Dakota['HIGH RISK PROPORTION']
X_ND_dummies=pd.get_dummies(X_ND)
X_ND_train,X_ND_test,y_ND_train,y_ND_test=train_test_split(X_ND_dummies,y_ND,random_state=0)
lin_ND=LinearRegression()
lin_ND.fit(X_ND_train,y_ND_train)
knn_ND=KNeighborsRegressor(n_neighbors=3)
knn_ND.fit(X_ND_train,y_ND_train)
lasso_ND=Lasso(alpha=0.01)
lasso_ND.fit(X_ND_train,y_ND_train)
ridge_ND=Ridge(alpha=0.01)
ridge_ND.fit(X_ND_train,y_ND_train)
dt_ND=DecisionTreeRegressor(random_state=0,max_leaf_nodes=2)
dt_ND.fit(X_ND_train,y_ND_train)
rf_ND=RandomForestRegressor(random_state=0,n_estimators=100)
rf_ND.fit(X_ND_train,y_ND_train)
print('NORTH DAKOTA')
print('ND Linear R^2 on Train Data:{:.3f}'.format(lin_ND.score(X_ND_train,y_ND_train)))
print('ND Linear R^2 on Test Data:{:.3f}'.format(lin_ND.score(X_ND_test,y_ND_test)))
print('ND KNN R^2 on Train Data:{:.3f}'.format(knn_ND.score(X_ND_train,y_ND_train)))
print('ND KNN R^2 on Test Data:{:.3f}'.format(knn_ND.score(X_ND_test,y_ND_test)))
print('ND Lasso R^2 on Train Data:{:.3f}'.format(lasso_ND.score(X_ND_train,y_ND_train)))
print('ND Lasso R^2 on Test Data:{:.3f}'.format(lasso_ND.score(X_ND_test,y_ND_test)))
print('ND Ridge R^2 on Train Data:{:.3f}'.format(ridge_ND.score(X_ND_train,y_ND_train)))
print('ND Ridge R^2 on Test Data:{:.3f}'.format(ridge_ND.score(X_ND_test,y_ND_test)))
print('ND Decision Tree R^2 on Train Data:{:.3f}'.format(dt_ND.score(X_ND_train,y_ND_train)))
print('ND Decision Tree R^2 on Test Data:{:.3f}'.format(dt_ND.score(X_ND_test,y_ND_test)))
print('ND Random Forest Regressor R^2 on Train Data:{:.3f}'.format(rf_ND.score(X_ND_train,y_ND_train)))
print('ND Random Forest Regressor R^2 on Test Data:{:.3f}'.format(rf_ND.score(X_ND_test,y_ND_test)))
print('')

RI=alcohol['STATE']=='RI'
Rhode_Island=alcohol[RI]
X_RI,y_RI=Rhode_Island.iloc[:,:-1],Rhode_Island['HIGH RISK PROPORTION']
X_RI_dummies=pd.get_dummies(X_RI)
X_RI_train,X_RI_test,y_RI_train,y_RI_test=train_test_split(X_RI_dummies,y_RI,random_state=0)
lin_RI=LinearRegression()
lin_RI.fit(X_RI_train,y_RI_train)
knn_RI=KNeighborsRegressor(n_neighbors=3)
knn_RI.fit(X_RI_train,y_RI_train)
lasso_RI=Lasso(alpha=0.01)
lasso_RI.fit(X_RI_train,y_RI_train)
ridge_RI=Ridge(alpha=0.01)
ridge_RI.fit(X_RI_train,y_RI_train)
dt_RI=DecisionTreeRegressor(random_state=0,max_leaf_nodes=2)
dt_RI.fit(X_RI_train,y_RI_train)
rf_RI=RandomForestRegressor(random_state=0,n_estimators=100)
rf_RI.fit(X_RI_train,y_RI_train)
print('RHODE ISLAND')
print('RI Linear R^2 on Train Data:{:.3f}'.format(lin_RI.score(X_RI_train,y_RI_train)))
print('RI Linear R^2 on Test Data:{:.3f}'.format(lin_RI.score(X_RI_test,y_RI_test)))
print('RI KNN R^2 on Train Data:{:.3f}'.format(knn_RI.score(X_RI_train,y_RI_train)))
print('RI KNN R^2 on Test Data:{:.3f}'.format(knn_RI.score(X_RI_test,y_RI_test)))
print('RI Lasso R^2 on Train Data:{:.3f}'.format(lasso_RI.score(X_RI_train,y_RI_train)))
print('RI Lasso R^2 on Test Data:{:.3f}'.format(lasso_RI.score(X_RI_test,y_RI_test)))
print('RI Ridge R^2 on Train Data:{:.3f}'.format(ridge_RI.score(X_RI_train,y_RI_train)))
print('RI Ridge R^2 on Test Data:{:.3f}'.format(ridge_RI.score(X_RI_test,y_RI_test)))
print('RI Decision Tree R^2 on Train Data:{:.3f}'.format(dt_RI.score(X_RI_train,y_RI_train)))
print('RI Decision Tree R^2 on Test Data:{:.3f}'.format(dt_RI.score(X_RI_test,y_RI_test)))
print('RI Random Forest Regressor R^2 on Train Data:{:.3f}'.format(rf_RI.score(X_RI_train,y_RI_train)))
print('RI Random Forest Regressor R^2 on Test Data:{:.3f}'.format(rf_RI.score(X_RI_test,y_RI_test)))
print('')

SC=alcohol['STATE']=='SC'
South_Carolina=alcohol[SC]
X_SC,y_SC=South_Carolina.iloc[:,:-1],South_Carolina['HIGH RISK PROPORTION']
X_SC_dummies=pd.get_dummies(X_SC)
X_SC_train,X_SC_test,y_SC_train,y_SC_test=train_test_split(X_SC_dummies,y_SC,random_state=0)
lin_SC=LinearRegression()
lin_SC.fit(X_SC_train,y_SC_train)
knn_SC=KNeighborsRegressor(n_neighbors=3)
knn_SC.fit(X_SC_train,y_SC_train)
lasso_SC=Lasso(alpha=0.01)
lasso_SC.fit(X_SC_train,y_SC_train)
ridge_SC=Ridge(alpha=0.01)
ridge_SC.fit(X_SC_train,y_SC_train)
dt_SC=DecisionTreeRegressor(random_state=0,max_leaf_nodes=2)
dt_SC.fit(X_SC_train,y_SC_train)
rf_SC=RandomForestRegressor(random_state=0,n_estimators=100)
rf_SC.fit(X_SC_train,y_SC_train)
print('SOUTH CAROLINA')
print('SC Linear R^2 on Train Data:{:.3f}'.format(lin_SC.score(X_SC_train,y_SC_train)))
print('SC Linear R^2 on Test Data:{:.3f}'.format(lin_SC.score(X_SC_test,y_SC_test)))
print('SC KNN R^2 on Train Data:{:.3f}'.format(knn_SC.score(X_SC_train,y_SC_train)))
print('SC KNN R^2 on Test Data:{:.3f}'.format(knn_SC.score(X_SC_test,y_SC_test)))
print('SC Lasso R^2 on Train Data:{:.3f}'.format(lasso_SC.score(X_SC_train,y_SC_train)))
print('SC Lasso R^2 on Test Data:{:.3f}'.format(lasso_SC.score(X_SC_test,y_SC_test)))
print('SC Ridge R^2 on Train Data:{:.3f}'.format(ridge_SC.score(X_SC_train,y_SC_train)))
print('SC Ridge R^2 on Test Data:{:.3f}'.format(ridge_SC.score(X_SC_test,y_SC_test)))
print('SC Decision Tree R^2 on Train Data:{:.3f}'.format(dt_SC.score(X_SC_train,y_SC_train)))
print('SC Decision Tree R^2 on Test Data:{:.3f}'.format(dt_SC.score(X_SC_test,y_SC_test)))
print('SC Random Forest Regressor R^2 on Train Data:{:.3f}'.format(rf_SC.score(X_SC_train,y_SC_train)))
print('SC Random Forest Regressor R^2 on Test Data:{:.3f}'.format(rf_SC.score(X_SC_test,y_SC_test)))
print('')

TN=alcohol['STATE']=='TN'
Tennessee=alcohol[TN]
X_TN,y_TN=Tennessee.iloc[:,:-1],Tennessee['HIGH RISK PROPORTION']
X_TN_dummies=pd.get_dummies(X_TN)
X_TN_train,X_TN_test,y_TN_train,y_TN_test=train_test_split(X_TN_dummies,y_TN,random_state=0)
lin_TN=LinearRegression()
lin_TN.fit(X_TN_train,y_TN_train)
knn_TN=KNeighborsRegressor(n_neighbors=3)
knn_TN.fit(X_TN_train,y_TN_train)
lasso_TN=Lasso(alpha=0.01)
lasso_TN.fit(X_TN_train,y_TN_train)
ridge_TN=Ridge(alpha=0.01)
ridge_TN.fit(X_TN_train,y_TN_train)
dt_TN=DecisionTreeRegressor(random_state=0,max_leaf_nodes=2)
dt_TN.fit(X_TN_train,y_TN_train)
rf_TN=RandomForestRegressor(random_state=0,n_estimators=100)
rf_TN.fit(X_TN_train,y_TN_train)
print('TENNESSEE')
print('TN Linear R^2 on Train Data:{:.3f}'.format(lin_TN.score(X_TN_train,y_TN_train)))
print('TN Linear R^2 on Test Data:{:.3f}'.format(lin_TN.score(X_TN_test,y_TN_test)))
print('TN KNN R^2 on Train Data:{:.3f}'.format(knn_TN.score(X_TN_train,y_TN_train)))
print('TN KNN R^2 on Test Data:{:.3f}'.format(knn_TN.score(X_TN_test,y_TN_test)))
print('TN Lasso R^2 on Train Data:{:.3f}'.format(lasso_TN.score(X_TN_train,y_TN_train)))
print('TN Lasso R^2 on Test Data:{:.3f}'.format(lasso_TN.score(X_TN_test,y_TN_test)))
print('TN Ridge R^2 on Train Data:{:.3f}'.format(ridge_TN.score(X_TN_train,y_TN_train)))
print('TN Ridge R^2 on Test Data:{:.3f}'.format(ridge_TN.score(X_TN_test,y_TN_test)))
print('TN Decision Tree R^2 on Train Data:{:.3f}'.format(dt_TN.score(X_TN_train,y_TN_train)))
print('TN Decision Tree R^2 on Test Data:{:.3f}'.format(dt_TN.score(X_TN_test,y_TN_test)))
print('TN Random Forest Regressor R^2 on Train Data:{:.3f}'.format(rf_TN.score(X_TN_train,y_TN_train)))
print('TN Random Forest Regressor R^2 on Test Data:{:.3f}'.format(rf_TN.score(X_TN_test,y_TN_test)))
print('')

VT=alcohol['STATE']=='VT'
Vermont=alcohol[VT]
X_VT,y_VT=Vermont.iloc[:,:-1],Vermont['HIGH RISK PROPORTION']
X_VT_dummies=pd.get_dummies(X_VT)
X_VT_train,X_VT_test,y_VT_train,y_VT_test=train_test_split(X_VT_dummies,y_VT,random_state=0)
lin_VT=LinearRegression()
lin_VT.fit(X_VT_train,y_VT_train)
knn_VT=KNeighborsRegressor(n_neighbors=3)
knn_VT.fit(X_VT_train,y_VT_train)
lasso_VT=Lasso(alpha=0.01)
lasso_VT.fit(X_VT_train,y_VT_train)
ridge_VT=Ridge(alpha=0.01)
ridge_VT.fit(X_VT_train,y_VT_train)
dt_VT=DecisionTreeRegressor(random_state=0,max_leaf_nodes=2)
dt_VT.fit(X_VT_train,y_VT_train)
rf_VT=RandomForestRegressor(random_state=0,n_estimators=100)
rf_VT.fit(X_VT_train,y_VT_train)
print('VERMONT')
print('VT Linear R^2 on Train Data:{:.3f}'.format(lin_VT.score(X_VT_train,y_VT_train)))
print('VT Linear R^2 on Test Data:{:.3f}'.format(lin_VT.score(X_VT_test,y_VT_test)))
print('VT KNN R^2 on Train Data:{:.3f}'.format(knn_VT.score(X_VT_train,y_VT_train)))
print('VT KNN R^2 on Test Data:{:.3f}'.format(knn_VT.score(X_VT_test,y_VT_test)))
print('VT Lasso R^2 on Train Data:{:.3f}'.format(lasso_VT.score(X_VT_train,y_VT_train)))
print('VT Lasso R^2 on Test Data:{:.3f}'.format(lasso_VT.score(X_VT_test,y_VT_test)))
print('VT Ridge R^2 on Train Data:{:.3f}'.format(ridge_VT.score(X_VT_train,y_VT_train)))
print('VT Ridge R^2 on Test Data:{:.3f}'.format(ridge_VT.score(X_VT_test,y_VT_test)))
print('VT Decision Tree R^2 on Train Data:{:.3f}'.format(dt_VT.score(X_VT_train,y_VT_train)))
print('VT Decision Tree R^2 on Test Data:{:.3f}'.format(dt_VT.score(X_VT_test,y_VT_test)))
print('VT Random Forest Regressor R^2 on Train Data:{:.3f}'.format(rf_VT.score(X_VT_train,y_VT_train)))
print('VT Random Forest Regressor R^2 on Test Data:{:.3f}'.format(rf_VT.score(X_VT_test,y_VT_test)))
print('')

VA=alcohol['STATE']=='VA'
Virginia=alcohol[VA]
X_VA,y_VA=Virginia.iloc[:,:-1],Virginia['HIGH RISK PROPORTION']
X_VA_dummies=pd.get_dummies(X_VA)
X_VA_train,X_VA_test,y_VA_train,y_VA_test=train_test_split(X_VA_dummies,y_VA,random_state=0)
lin_VA=LinearRegression()
lin_VA.fit(X_VA_train,y_VA_train)
knn_VA=KNeighborsRegressor(n_neighbors=3)
knn_VA.fit(X_VA_train,y_VA_train)
lasso_VA=Lasso(alpha=0.01)
lasso_VA.fit(X_VA_train,y_VA_train)
ridge_VA=Ridge(alpha=0.01)
ridge_VA.fit(X_VA_train,y_VA_train)
dt_VA=DecisionTreeRegressor(random_state=0,max_leaf_nodes=2)
dt_VA.fit(X_VA_train,y_VA_train)
rf_VA=RandomForestRegressor(random_state=0,n_estimators=100)
rf_VA.fit(X_VA_train,y_VA_train)
print('VIRGINIA')
print('VA Linear R^2 on Train Data:{:.3f}'.format(lin_VA.score(X_VA_train,y_VA_train)))
print('VA Linear R^2 on Test Data:{:.3f}'.format(lin_VA.score(X_VA_test,y_VA_test)))
print('VA KNN R^2 on Train Data:{:.3f}'.format(knn_VA.score(X_VA_train,y_VA_train)))
print('VA KNN R^2 on Test Data:{:.3f}'.format(knn_VA.score(X_VA_test,y_VA_test)))
print('VA Lasso R^2 on Train Data:{:.3f}'.format(lasso_VA.score(X_VA_train,y_VA_train)))
print('VA Lasso R^2 on Test Data:{:.3f}'.format(lasso_VA.score(X_VA_test,y_VA_test)))
print('VA Ridge R^2 on Train Data:{:.3f}'.format(ridge_VA.score(X_VA_train,y_VA_train)))
print('VA Ridge R^2 on Test Data:{:.3f}'.format(ridge_VA.score(X_VA_test,y_VA_test)))
print('VA Decision Tree R^2 on Train Data:{:.3f}'.format(dt_VA.score(X_VA_train,y_VA_train)))
print('VA Decision Tree R^2 on Test Data:{:.3f}'.format(dt_VA.score(X_VA_test,y_VA_test)))
print('VA Random Forest Regressor R^2 on Train Data:{:.3f}'.format(rf_VA.score(X_VA_train,y_VA_train)))
print('VA Random Forest Regressor R^2 on Test Data:{:.3f}'.format(rf_VA.score(X_VA_test,y_VA_test)))
print('')

WV=alcohol['STATE']=='WV'
West_Virginia=alcohol[WV]
X_WV,y_WV=West_Virginia.iloc[:,:-1],West_Virginia['HIGH RISK PROPORTION']
X_WV_dummies=pd.get_dummies(X_WV)
X_WV_train,X_WV_test,y_WV_train,y_WV_test=train_test_split(X_WV_dummies,y_WV,random_state=0)
lin_WV=LinearRegression()
lin_WV.fit(X_WV_train,y_WV_train)
knn_WV=KNeighborsRegressor(n_neighbors=3)
knn_WV.fit(X_WV_train,y_WV_train)
lasso_WV=Lasso(alpha=0.01)
lasso_WV.fit(X_WV_train,y_WV_train)
ridge_WV=Ridge(alpha=0.01)
ridge_WV.fit(X_WV_train,y_WV_train)
dt_WV=DecisionTreeRegressor(random_state=0,max_leaf_nodes=2)
dt_WV.fit(X_WV_train,y_WV_train)
rf_WV=RandomForestRegressor(random_state=0,n_estimators=100)
rf_WV.fit(X_WV_train,y_WV_train)
print('WEST VIRGINIA')
print('WV Linear R^2 on Train Data:{:.3f}'.format(lin_WV.score(X_WV_train,y_WV_train)))
print('WV Linear R^2 on Test Data:{:.3f}'.format(lin_WV.score(X_WV_test,y_WV_test)))
print('WV KNN R^2 on Train Data:{:.3f}'.format(knn_WV.score(X_WV_train,y_WV_train)))
print('WV KNN R^2 on Test Data:{:.3f}'.format(knn_WV.score(X_WV_test,y_WV_test)))
print('WV Lasso R^2 on Train Data:{:.3f}'.format(lasso_WV.score(X_WV_train,y_WV_train)))
print('WV Lasso R^2 on Test Data:{:.3f}'.format(lasso_WV.score(X_WV_test,y_WV_test)))
print('WV Ridge R^2 on Train Data:{:.3f}'.format(ridge_WV.score(X_WV_train,y_WV_train)))
print('WV Ridge R^2 on Test Data:{:.3f}'.format(ridge_WV.score(X_WV_test,y_WV_test)))
print('WV Decision Tree R^2 on Train Data:{:.3f}'.format(dt_WV.score(X_WV_train,y_WV_train)))
print('WV Decision Tree R^2 on Test Data:{:.3f}'.format(dt_WV.score(X_WV_test,y_WV_test)))
print('WV Random Forest Regressor R^2 on Train Data:{:.3f}'.format(rf_WV.score(X_WV_train,y_WV_train)))
print('WV Random Forest Regressor R^2 on Test Data:{:.3f}'.format(rf_WV.score(X_WV_test,y_WV_test)))
print('')

WY=alcohol['STATE']=='WY'
Wyoming=alcohol[WY]
X_WY,y_WY=Wyoming.iloc[:,:-1],Wyoming['HIGH RISK PROPORTION']
X_WY_dummies=pd.get_dummies(X_WY)
X_WY_train,X_WY_test,y_WY_train,y_WY_test=train_test_split(X_WY_dummies,y_WY,random_state=0)
lin_WY=LinearRegression()
lin_WY.fit(X_WY_train,y_WY_train)
knn_WY=KNeighborsRegressor(n_neighbors=3)
knn_WY.fit(X_WY_train,y_WY_train)
lasso_WY=Lasso(alpha=0.01)
lasso_WY.fit(X_WY_train,y_WY_train)
ridge_WY=Ridge(alpha=0.01)
ridge_WY.fit(X_WY_train,y_WY_train)
dt_WY=DecisionTreeRegressor(random_state=0,max_leaf_nodes=2)
dt_WY.fit(X_WY_train,y_WY_train)
rf_WY=RandomForestRegressor(random_state=0,n_estimators=100)
rf_WY.fit(X_WY_train,y_WY_train)
print('WYOMING')
print('WY Linear R^2 on Train Data:{:.3f}'.format(lin_WY.score(X_WY_train,y_WY_train)))
print('WY Linear R^2 on Test Data:{:.3f}'.format(lin_WY.score(X_WY_test,y_WY_test)))
print('WY KNN R^2 on Train Data:{:.3f}'.format(knn_WY.score(X_WY_train,y_WY_train)))
print('WY KNN R^2 on Test Data:{:.3f}'.format(knn_WY.score(X_WY_test,y_WY_test)))
print('WY Lasso R^2 on Train Data:{:.3f}'.format(lasso_WY.score(X_WY_train,y_WY_train)))
print('WY Lasso R^2 on Test Data:{:.3f}'.format(lasso_WY.score(X_WY_test,y_WY_test)))
print('WY Ridge R^2 on Train Data:{:.3f}'.format(ridge_WY.score(X_WY_train,y_WY_train)))
print('WY Ridge R^2 on Test Data:{:.3f}'.format(ridge_WY.score(X_WY_test,y_WY_test)))
print('WY Decision Tree R^2 on Train Data:{:.3f}'.format(dt_WY.score(X_WY_train,y_WY_train)))
print('WY Decision Tree R^2 on Test Data:{:.3f}'.format(dt_WY.score(X_WY_test,y_WY_test)))
print('WY Random Forest Regressor R^2 on Train Data:{:.3f}'.format(rf_WY.score(X_WY_train,y_WY_train)))
print('WY Random Forest Regressor R^2 on Test Data:{:.3f}'.format(rf_WY.score(X_WY_test,y_WY_test)))
```
